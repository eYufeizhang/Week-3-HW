```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Load the penguins dataset
penguins = sns.load_dataset('penguins')
# Dropping rows with NaN values in the 'flipper_length_mm' column
penguins = penguins.dropna(subset=['flipper_length_mm']).
# Extract the 'flipper_length_mm' column
flipper_lengths = penguins['flipper_length_mm']
# Calculate the statistics
mean_val = flipper_lengths.mean()
median_val = flipper_lengths.median()
min_val = flipper_lengths.min()
max_val = flipper_lengths.max()
# Plot the histogram
plt.figure(figsize=(10,6))
sns.histplot(flipper_lengths, bins=20, kde=True, color='skyblue')
# Annotating the mean, median, and range
plt.axvline(mean_val, color='r', linestyle='--', label=f'Mean: {mean_val:.2f} mm')
plt.axvline(median_val, color='g', linestyle='--', label=f'Median: {median_val:.2f} mm')
plt.axvline(min_val, color='b', linestyle='--', label=f'Min: {min_val} mm')
plt.axvline(max_val, color='b', linestyle='--', label=f'Max: {max_val} mm')
# Adding labels and title
plt.title('Histogram of Flipper Length (mm) in Penguins with Annotations')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.legend()
# Show the plot
plt.show()
```
